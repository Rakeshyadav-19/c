```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
#from sklearn.datasets import fetch_brain_tumor_dataset
```

```python
df= pd.read_csv('brain_tumor_dataset.csv')
df
```

```python
print(df.columns)
df = df.rename(columns={df.columns[2]: 'Size'})
df = df.rename(columns={df.columns[4]: 'Age'})
df
```

```python
x= df.Size
y= df.Age
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train = x_train.values.reshape(-1, 1)
x_test = x_test.values.reshape(-1, 1)

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_predict= model.predict(x_test)
```

```python
mse= mean_squared_error(y_test, y_predict)
r2= r2_score(y_test, y_predict)
print(mse,r2)
```

```python
plt.scatter(y_test, y_predict, edgecolors=(0, 0, 0))
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
plt.xlabel('Measured')
plt.ylabel('Predicted values')
plt.title('Measured vs Predicted value')
plt.show()
```
